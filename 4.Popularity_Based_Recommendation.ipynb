```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# reading ratings file:
colms=['user_id', 'movie_id', 'rating', 'unix_timestamp']
ratings=pd.read_csv('ml-100k/u.data',sep='\t', names= colms,encoding='latin-1')
ratings
```

```python
n_users=ratings.user_id.unique().shape[0]
n_items=ratings.movie_id.unique().shape[0]
```

```python
print("The number of user:",n_users)
print("The number of n_items:",n_items)
```

```python
datama=ratings.pivot_table(index='user_id',columns='movie_id',values='rating')
datamatri=datama
datamatrix=datamatri.replace(np.nan,0)
datamatrix
```

```python
#Read the movie Dataset
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
item=pd.read_csv('ml-100k/u.item',sep='|', names=i_cols, encoding='latin-1')
```

```python
movie=item[['Action', 'Adventure',
    'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
    'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']]
movie
```

```python
genric_ids = movie[movie['Action'] == 1].index
genric_ids
```

```python
gen_rating=datamatrix[genric_ids]
genric_rating=gen_rating.T
genric_rating
```

```python
mean=genric_rating.mean(axis=1)
avg = pd.DataFrame(mean)
avg.columns = ["avg_rating"]
avg=avg.sort_values(by='avg_rating',ascending=False).head(5)
avg
```

```python
movie_title=[]
for high_rate in avg.index:
    mov_ti=item[item['movie id']==high_rate]['movie title'].values
    movie_title.append(mov_ti)
```

```python
#movie title as one list
list = np.concatenate(movie_title)
print(list)
```

```python
avg['movie_title']=list
avg
```

```python
def popular_based(genric):
    genric_ids = movie[movie[genric] == 1].index
    gen_rating=datamatrix[genric_ids]
    genric_rating=gen_rating.T
    mean=genric_rating.mean(axis=1)
    avg = pd.DataFrame(mean)
    avg.columns = ["avg_rating"]
    avg=avg.sort_values(by='avg_rating',ascending=False).head(5)
    movie_title=[]
    for high_rate in avg.index:
        mov_ti=item[item['movie id']==high_rate]['movie title'].values
        movie_title.append(mov_ti)
    list = np.concatenate(movie_title)
    avg['movie_title']=list
    avg
    return avg   
        
        
```

```python
popular_based('Mystery')
```
